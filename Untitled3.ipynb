```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('salary_data.csv')

media_salarial_por_pais = df.groupby('country_name')['median_salary'].mean()

paises_com_menor_media = media_salarial_por_pais.sort_values().head(50)

nomes_paises = paises_com_menor_media.index

plt.figure(figsize=(12, 6))
plt.bar(nomes_paises, paises_com_menor_media)
plt.title('Média Salarial dos 50 Países com Menor Média')
plt.xlabel('País')
plt.ylabel('Média Salarial')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
